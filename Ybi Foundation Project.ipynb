Data Preprocessing

# Handling categorical features
After exploring the dataset, I noticed the need to convert certain categorical variables into dummy variables using the `get_dummies()` function. Although the dataset does not contain any string-based features, the following columns are categorical in nature and should be encoded accordingly: `'sex'`, `'cp'`, `'fbs'`, `'restecg'`, `'exang'`, `'slope'`, `'ca'`, and `'thal'`.

For example, consider the `'sex'` column. Although it contains only binary values (0 and 1), leaving it as-is might lead the algorithm to interpret the values as ordinal  implying that one gender is greater than the other. Since gender is a nominal variable, such an interpretation would be misleading. Therefore, it's important to treat these features as categorical and convert them into dummy variables to avoid introducing unintended ordinal relationships.


```python
dataset = pd.get_dummies(df, columns=['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

## Feature Scaling

```python
dataset.columns
```

```python
from sklearn.preprocessing import StandardScaler
standScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standScaler.fit_transform(dataset[columns_to_scale])

dataset.head()
```

```python
# Splitting the dataset into dependent and independent features
X = dataset.drop('target', axis=1)
y = dataset['target']
```

# Model Building

## KNeighbors Classifier Model

```python
# Importing essential libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
# Finding the best accuracy for knn algorithm using cross_val_score 
knn_scores = []
for i in range(1, 21):
  knn_classifier = KNeighborsClassifier(n_neighbors=i)
  cvs_scores = cross_val_score(knn_classifier, X, y, cv=10)
  knn_scores.append(round(cvs_scores.mean(),3))
```

```python
# Plotting the results of knn_scores
plt.figure(figsize=(20,15))
plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
plt.show()
```
